# Segmenting and Clustering Neighborhoods in Toronto

```python
#import libraies
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup # this module helps in web scrapping.
import requests  # this module helps us to download a web page
```

# Get the neighborhood data using beautiful soup

```python
url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
data  = requests.get(url).text
```

```python
soup = BeautifulSoup(data,"html5lib")  # create a soup object using the variable 'data'
```

```python
tables = soup.find('table') # in html table is represented by the tag <table>
```

```python
toronto_data = pd.DataFrame(columns=["PostalCode", "Borough", "Neighbourhood"])

for row in tables.tbody.find_all("tr"):
    col = row.find_all("td")
    if (col != []):
        PostalCode = col[0].text.strip()
        Borough = col[1].text.strip()
        Neighbourhood = col[2].text.strip()
        toronto_data = toronto_data.append({"PostalCode":PostalCode, "Borough":Borough, "Neighbourhood":Neighbourhood}, ignore_index=True)

toronto_data
```

```python
toronto_data["Borough"]=toronto_data["Borough"].replace(to_replace ="Not assigned", value =np.nan) 
pd.set_option('display.max_rows', None)
toronto_data
```

```python
toronto_data=toronto_data.dropna()
```

```python
toronto_data.reset_index(drop=True, inplace=True)
toronto_data
```

# generate maps to visualize your neighborhoods and how they cluster together

```python
geo_url = "https://cocl.us/Geospatial_data"

geo_df = pd.read_csv(geo_url)
geo_df.rename(columns={'Postal Code': 'PostalCode'}, inplace=True)
geo_df.head()
```

```python
df = pd.merge(toronto_data, geo_df, on='PostalCode')
df.head()
```

```python
#Create list with the Boroughs (to be used later)
boroughs = df['Borough'].unique().tolist()
```

```python
#Obtain the coordinates from the dataset itself, just averaging Latitude/Longitude of the current dataset 
lat_toronto = df['Latitude'].mean()
lon_toronto = df['Longitude'].mean()
print('The geographical coordinates of Toronto are {}, {}'.format(lat_toronto, lon_toronto))
```

```python
# This will color categorize each borough
borough_color = {}
for borough in boroughs:
    borough_color[borough]= '#%02X%02X%02X' % tuple(np.random.choice(range(256), size=3)) #Random color
```

```python
import folium
map_toronto = folium.Map(location=[lat_toronto, lon_toronto], zoom_start=10.5)

# add markers to map
for lat, lng, borough, neighborhood in zip(df['Latitude'], 
                                           df['Longitude'],
                                           df['Borough'], 
                                           df['Neighbourhood']):
    label_text = borough + ' - ' + neighborhood
    label = folium.Popup(label_text)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color=borough_color[borough],
        fill_color=borough_color[borough],
        fill_opacity=0.8).add_to(map_toronto)  
    
map_toronto
```
